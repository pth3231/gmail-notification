```python
import sys
import os

# Add parent directory to path so we can import modules
sys.path.insert(0, os.path.abspath(os.path.join(os.getcwd(), '..')))

import gmail_service


print("Step 1: Fetching unread emails from Gmail...")

result = gmail_service.fetch_unread_emails(max_results=3)
    
if not result:
    print("No emails to process.")

messages, service = result
print(f"Fetched {len(messages)} unread email(s)\n")
print(messages)
```

```python
import email_processor
import cleaning
import re

print("Step 2: Extracting email details...")

extracted_data = []
    
for msg in messages:
    msg_detail = service.users().messages().get(
        userId='me', 
        id=msg['id'], 
        format='full'
    ).execute()
    
    payload = msg_detail['payload']
    headers = payload.get('headers', [])
    
    # Extract headers
    sender = next((header['value'] for header in headers if header['name'] == 'From'), "Unknown")
    subject = next((header['value'] for header in headers if header['name'] == 'Subject'), "No Subject")
    
    # Extract body
    body_raw = cleaning.extract_email_body(payload)
    
    # Apply full cleaning (this has all the regex replacements)
    body_cleaned = cleaning.clean_email_content(body_raw)
    print(f"--- Cleaned body preview (first 500 chars) ---")
    print(body_cleaned[:500])
    print(f"Cleaned body length: {len(body_cleaned)} chars\n")
    
    extracted_data.append({
        "sender": sender,
        "subject": subject,
        "body": body_cleaned
    })
    
    print(f"Loaded: {subject}")
    print("=" * 70)

if not extracted_data:
    print("No email data extracted.")

print(f"\nTotal emails extracted: {len(extracted_data)}")
for idx, email in enumerate(extracted_data, 1):
    print(f"\n{idx}. {email['subject']}")
    print(f"   From: {email['sender']}")
    print(f"   Body preview: {email['body'][:200]}...")
```
